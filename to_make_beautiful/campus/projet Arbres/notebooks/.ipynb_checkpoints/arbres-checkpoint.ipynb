```python
#shows all genera of trees presented in the dataframe

len(set(genre))-1 # -1 because genre list has '' values by default
```

```python
# creates a list of all species of trees ever planted in grenob

with open('trees.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    espece = [str(i[14]) for i in list(reader)[1:]]

espece = list(zip(genre,espece))
espece = [i for i in espece 
           if (i[0] != '') 
           and (i[1] != '')]

espece = [i[0]+' '+i[1] for i in espece]
#espece = set(espece)
```

```python
# creates a list of counts for each species

number_of_espece = [espece.count(i) for i in espece]
```

```python
#creates a dict for all the species in dataset

espece_dict = dict(zip(espece,number_of_espece))
```

```python
#creates a dict for genre and its count 

genre_dict = dict(zip(genre,[genre.count(i) for i in genre]))
del genre_dict['']
```

```python
#shows the most popular genre d'arbre ever planted in grenob

print([i for i in genre_dict.keys() if (genre_dict[i] == max(genre_dict.values()))][0])
```

```python
#shows a list of unique genres in grenob (planted only once)

print([i for i in genre_dict.keys() if (genre_dict[i] == 1)])
```

```python
#returns a dict of genres and its counts in a descending order

{gen: esp for gen, esp in sorted(genre_dict.items(), key=lambda item: item[1], reverse=True)}
```

```python
#returns the proportion of the genres of trees in grenob  

def proportion_du_genre(genre = None):
    lst = [i[1] / sum(genre_dict.values()) for i in genre_dict.items()]

    if genre != None:
        for i in genre_dict.items():
            print(i[0]+':', round((i[1] / sum(genre_dict.values()))*100,2),'%')        
        
    else:
        print(round(max(lst)*100, 2),'%',sep = '')
```

```python
#proportion of the most common tree genre in grenob in %

proportion_du_genre()
```

```python
# returns numbers of smth planted each year as a dictionary
# takes list of tuples (year,'smth') as an argument

def combine(list_of_tuples):
    result_dict = dict()
    for key, value in list_of_tuples:
        # Add the key-value pair to the dictionary
        if key in result_dict.keys():
            result_dict[key].add(value)
        else:
            result_dict[key] = {value}

    return result_dict
```

```python
#gives a dict of (year: set of genres planted that year) 

year_genre_dict = combine(year_genre())
```

```python
divers_genre_year = [len(i) for i in year_genre_dict.values()]

x = [int(i) for i in year_genre_dict.keys()]
```

```python
# shows a change of number of genres planted since 1950 in Grenoble 

import numpy as np

y=divers_genre_year

coef = np.polyfit(x,y,1)  # takes in x and returns an estimate for y
poly1d_fn = np.poly1d(coef)

fig = plt.figure(figsize=(5,3))

plt.plot(x,y, 'bo', x, poly1d_fn(x),'--k')
plt.xlabel('Year')
plt.ylabel('Number of genres')
plt.tight_layout()
plt.show()
```

```python
#gives a dict of (year: set of genres planted that year) 

def year_espece():
    lst = list(zip(annedeplant,espece))
    lst = [i for i in lst 
           if (i[0] != '') 
           and (i[1] != '')]
    lst.sort()
    return lst

year_espece_dict = combine(year_espece())

divers_espece_year = [len(i) for i in year_espece_dict.values()]

x = [int(i) for i in year_espece_dict.keys()]
```

```python
# shows a change of number of species planted since 1950 in Grenoble 

import numpy as np

y=divers_espece_year

coef = np.polyfit(x,y,1)  # takes in x and returns an estimate for y
poly1d_fn = np.poly1d(coef)

fig = plt.figure(figsize=(5,3))

plt.plot(x,y, 'ro', x, poly1d_fn(x),'--k')
plt.xlabel('Year')
plt.ylabel('Number of species')
plt.tight_layout()
plt.show()
```

```python
species = pd.concat([data_raw['GENRE_BOTA'],data_raw['ESPECE']],axis=1)
species = species.dropna()
species_good = species[species['GENRE_BOTA'] != 'Betula']
species_good = species_good[species_good['GENRE_BOTA'] != 'Castanea']
species_good = species_good[species_good['GENRE_BOTA'] != 'Cupressus']
species_good = species_good[species_good['GENRE_BOTA'] != 'Fraxinus']
species_good = species_good[species_good['GENRE_BOTA'] != 'Olea']
species_good = species_good[species_good['GENRE_BOTA'] != 'Platanus']
species_good = species_good[species_good['GENRE_BOTA'] != 'Aesculus']
species_good
```

```python
data_bad_trees = pd.read_csv('bad.csv')
data_bad_trees.info()
```

```python
def separation(genre,espece):
    for i in range(len(species_good)):
        if ((species_good.iloc[i, 0] != genre)&(species_good.iloc[i,1] != espece)):
            return 1
        else:
            return 0
```

```python
merged_df = species_good.merge(data_bad_trees, how='left', left_on=['GENRE_BOTA', 'ESPECE'], right_on=['Genre', ' espece'])
merged_df
```

```python
species_good['good'] = merged_df['Genre'].isna().astype(int)
species_good.dropna()
```

```python
species_good.value_counts().to_frame('count').reset_index()
```

```python
(species_good.value_counts().sum()/species.value_counts().sum())*100
```

```python
species_goodonly = species_good.value_counts().to_frame('count').reset_index()
```

```python
species_goodonly['species'] = species_goodonly.GENRE_BOTA+' '+species_goodonly.ESPECE
```

```python
species_goodonly
```

```python
task = species_goodonly.groupby('GENRE_BOTA').sum().sort_values(by='count',ascending=False).drop('good',axis=1)
```

```python
task['proportion'] = (task['count']/task['count'].sum())*100
task
```

```python
to_plant = task.copy()
```

```python
to_plant = to_plant[to_plant['proportion']<15]
```

```python
to_plant = to_plant.reset_index()
to_plant
```

```python
species_goodonly
```

```python
species_goodonly_noacer = species_goodonly[species_goodonly['GENRE_BOTA']!='Acer']
```

```python
species_goodonly_noacer
```

```python
def number_to_plant(plant, planted = 0):
    x1 = species_goodonly_noacer['count'][species_goodonly_noacer['species']==plant]
    x2 = to_plant['count'][to_plant['GENRE_BOTA']==plant.split(' ')[0]]
    sum1 = to_plant['count'].sum() + planted
    lst = []
    for x in range(1,3000):
        prop1 = ((x+x1)/(sum1+x))*100
        prop2 = ((x+x2)/(sum1+x))*100
        if ((prop1 <= 10).any() 
            & (prop2 <= 15).any()):
            lst.append(x)
    return max(lst)
```
