## Visualize Data

### <font color='purple'> Quels sont les types de variables ? D√©crivez-les √† l'aide d'un graphique. </font>

```python
types = pd.DataFrame(df_energy.dtypes,columns=['type'])
types
```

```python
sns.countplot(types,x='type')
```

### <font color='purple'> Quelle est l'√©volution de la consommation ? </font>

```python
df_energy['DateTime'] = pd.to_datetime(df_energy['DateTime'])
```

```python
df_energy = df_energy.set_index('DateTime')
```

```python
df_energy_Y = df_energy.resample('Y').mean()
df_energy_Y.index = np.arange(2012,2019,1)
```

```python
sns.barplot(df_energy_Y, x = df_energy_Y.index, y = df_energy_Y['Consommation'])
plt.yscale('log')
plt.show
```

```python
sns.lineplot(df_energy_Y, x = df_energy_Y.index,y = 'Consommation')
plt.fill_between(df_energy_Y.index, df_energy_Y['Consommation'], alpha=0.3)
plt.ylim(50000,60000)
plt.xlim(2012,2018)
```

### <font color='purple'> Quand a-t-on a un exc√©dant / un d√©ficit de production ? </font>

```python
df_energy_Y
```

```python
df_energy_Y['Production'] = df_energy_Y.drop(['Consommation','Pompage','Ech. physiques'],axis=1).apply(lambda x: x.sum(),axis=1)
```

```python
sns.lineplot(df_energy_Y, x = df_energy_Y.index,y = 'Production',label='Production')
plt.fill_between(df_energy_Y.index, df_energy_Y['Production'], alpha=0.3)
sns.lineplot(df_energy_Y, x = df_energy_Y.index,y = 'Consommation',label='Consommation')
plt.fill_between(df_energy_Y.index, df_energy_Y['Consommation'], alpha=0.5)
plt.ylim(50000,70000)
plt.xlim(2012,2018)
plt.legend()
plt.ylabel('')
plt.xlabel('')
```

### <font color='purple'> Y a-t-il une corr√©lation entre les diff√©rents types de production ? </font>

```python
corr = df_energy_Y.drop(['Consommation','Pompage','Ech. physiques','Production'],axis=1).corr()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', vmax=1, vmin=-1)
```

#### Table

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 4))
axes = axes.flatten()

# sns.barplot(df_energy_Y, x = df_energy_Y.index, y = df_energy_Y['Consommation'],ax = axes[0])
# axes[0].set_yscale('log')

sns.lineplot(df_energy_Y, x = df_energy_Y.index,y = 'Consommation',ax = axes[0])
axes[0].fill_between(df_energy_Y.index, df_energy_Y['Consommation'], alpha=0.3)
axes[0].set_ylim(50000,60000)
axes[0].set_xlim(2012,2018)
axes[0].set_xlabel('')

sns.lineplot(df_energy_Y, x = df_energy_Y.index,y = 'Production',label='Production',ax = axes[1])
axes[1].fill_between(df_energy_Y.index, df_energy_Y['Production'], alpha=0.3)
sns.lineplot(df_energy_Y, x = df_energy_Y.index,y = 'Consommation',ax = axes[1],label='Consommation')
axes[1].fill_between(df_energy_Y.index, df_energy_Y['Consommation'], alpha=0.3)
axes[1].set_ylim(50000,70000)
axes[1].set_xlim(2012,2018)
axes[1].set_ylabel('')
axes[1].set_xlabel('')
axes[1].legend()

sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', vmax=1, vmin=-1,ax = axes[2])

row_titles = ['Energy', 'Energy']
column_titles = ['Question 1', 'Question 2', 'Question 3', 'Question 4', 'Question 5']

for i, ax in enumerate(axes[::5]):
    ax.set_ylabel(row_titles[i], size='large', labelpad=20)

for i, ax in enumerate(axes[:5]):
    ax.set_title(column_titles[i], size='large', pad=10)

plt.savefig('./energy.png')
plt.show()
```
